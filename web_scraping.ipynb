```python
import requests
from bs4 import BeautifulSoup
import csv

response = requests.get("https://www.wetter.com/deutschland/EUDE.html").content
soup = BeautifulSoup(response, 'html.parser')
print(soup.prettify())

table = soup.find("table", {"class": "[ one-whole ][ table table--rows table--compact no-border ][ mb0 ]"})


thead = table.find("thead").find_all("th")
table_head = [th.text for th in thead]
table_body=table.find("tbody").find_all("tr")

with open("report.cvs", "w") as csv_file:
  csv_write= csv.writer(csv_file)
  csv_write.writerow(table_head)

  for tr in table_body:
    table_data = [td.text.strip() for td in tr.find_all('td')]
    csv_write.writerow(table_data)
```
